### Select K
Identify a suitable choice of K for the given data set from the above exercise.

K = 2 is a suitable choice as it has the least distnace variation

```python
km = KMeans(n_clusters=2)
def k_mean_distance(data, cx, cy, i_centroid, cluster_labels):
        distances = [np.sqrt((x-cx)**2+(y-cy)**2) for (x, y) in data[cluster_labels == i_centroid]]
        return distances
clusters=km.fit_predict(x)
centroids = km.cluster_centers_

distances = []
for i, (cx, cy) in enumerate(centroids):
    mean_distance = k_mean_distance(x, cx, cy, i, clusters)
    distances.append(mean_distance)

arr1 = np.asarray(distances[1])
arr0 = np.asarray(distances[0])
di_2 = np.sum(arr1) + np.sum(arr0)
y_kmeans = km.predict(x)
plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, alpha=0.5);
```
